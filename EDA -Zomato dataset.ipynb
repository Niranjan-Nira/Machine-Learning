```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
f_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
f_df.head(2)
```

```python
country_names=f_df.Country.value_counts().index
```

```python
country_val=f_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.1f%%')
```

Observation: Zomato maximum records are from india.After the united states and then united kingdom

```python
f_df.columns
```

```python
ratings = f_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observation: when the rating is between 4.5 to 4.9--->Excellent
when the rating is between 4.0 to 3.4--->very good
when the rating is between 3.5 to 3.9--->good
when the rating is between 3.0.5 to 3.4--->average

```python
ratings.head()
```

```python
sns.barplot(x = 'Aggregate rating',y = 'Rating Count',data=ratings)
```

```python
sns.barplot(x = 'Aggregate rating',y = 'Rating Count',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

Observation:
not rated count is very high
maximun number of rating is between 2.5 to 3.5

```python
sns.countplot(x = 'Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
